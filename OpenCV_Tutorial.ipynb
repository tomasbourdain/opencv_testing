```python
import cv2
import os
import matplotlib.pyplot as plt
```

## Read In Image

```python
img_path = os.path.join('OpenCV','data','images','demon1.jpg')
```

```python
img_path
```

```python
img = cv2.imread(img_path)
```

```python
img.shape
```

## Render Image

```python
plt.imshow(img)
```

 This happens because of the way OpenCV works with the color channels and how matplotlib works with them
 <p>
 Matplotlib expects that the image would be in the format RGB, but OpenCV uses format BGR

```python
recolor = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(recolor)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray)
```

## Write Out Image

```python
cv2.imwrite('grayimage.jpg', gray)
```

```python
output_path = os.path.join('OpenCV','data','images','outputgray.jpg')
```

```python
cv2.imwrite(output_path, gray)
```
