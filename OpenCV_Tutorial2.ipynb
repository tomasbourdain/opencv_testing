# OpenCV - Connecting to My Webcam

```python
import cv2
import os
import matplotlib.pyplot as plt
```

## Connect to Webcam

```python
cap = cv2.VideoCapture(0)
```

```python
# Get a frame from the capture device
ret, frame = cap.read()
```

```python
plt.imshow(frame)
```

```python
# Releases capture back into the wild
cap.release()
```

```python
def take_photo():
    cap = cv2.VideoCapture(0)
    ret, frame = cap.read()
    cv2.imwrite('webcamphoto.jpg', frame)
    cap.release()
```

```python
take_photo()
```

## Rendering in Real-Time

```python
# Connect to webcam
cap = cv2.VideoCapture(0)
#Loop through every frame until we close our webcam
while cap.isOpened():
    ret, frame = cap.read()

    # Show image 
    cv2.imshow('Webcam', frame)

    # Check if q has been hit and stops the loop
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
# Releases webcam
cap.release()
# Closes the frame
cv2.destroyAllWindows()
```
