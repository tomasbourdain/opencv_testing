# OpenCV - Detect Edges

```python
import cv2
import os
import matplotlib.pyplot as plt
```

## Load image

```python
img_path = os.path.join('data','images','demon.jpg')
```

```python
# Read in a image
# -1, cv2.IMREAD_COLOR 
# 0, cv2.IMREAD_GRAYSCALE
# 1, cv2.IMREAD_UNCHANGED
img = cv2.imread(img_path)
img
```

```python
plt.imshow(img)
```

```python
# Resize image 
resized_img = cv2.resize(img, (int(img.shape[1]/2), int(img.shape[0]/2)))
#View image using OpenCV
cv2.imshow('Frame View', resized_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Edge Detection

The tighter the threshold the less number of edges, the wider the threshold the more number of edges

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# First step is to remove the noise in the image with a 5x5 Gaussian Filter
blur = cv2.GaussianBlur(gray, (5,5), 0)
# Apply Canny
canny = cv2.Canny(blur, threshold1=220, threshold2=240)
```

```python
# Resize image 
resized_img = cv2.resize(canny, (int(img.shape[1]/2), int(img.shape[0]/2)))
#View image using OpenCV
cv2.imshow('Frame View', resized_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Export Edges

```python
cv2.imwrite('edgesdemon.jpg', canny)
```
